## 2. Apply transform specs

One harmonized frame (`h_*`) per cohort, keyed by CDE.

```python
# ===== AoU =====
h_AoU = pd.DataFrame(index=raw_AoU.index)

# Biological sex  ·  adopt (categorical recode)
_map = {'1': 'Male', '2': 'Female'}
h_AoU["Sex Assigned At Birth"] = raw_AoU["SEX"].astype(str).map(_map)

# Age at enrollment  ·  adopt (copy)
h_AoU["Age At Enrollment (years)"] = raw_AoU["AGE"]

# Diastolic blood pressure  ·  refine (copy)
h_AoU["Diastolic Blood Pressure (mmHg)"] = raw_AoU["dbp"]

# Diabetes diagnosis  ·  refine (categorical recode)  # 1 source code(s) unmapped → NaN
_map = {'1': 'Yes', '0': 'No'}
h_AoU["Diabetes Mellitus Indicator"] = raw_AoU["diabetes"].astype(str).map(_map)

# Height (cm)  ·  adopt (copy)
h_AoU["Standing Height (cm)"] = raw_AoU["height"]
```

```python
# ===== CLSA =====
h_CLSA = pd.DataFrame(index=raw_CLSA.index)

# Biological sex  ·  adopt (categorical recode)
_map = {'M': 'Male', 'F': 'Female'}
h_CLSA["Sex Assigned At Birth"] = raw_CLSA["SEX_ASK_TRM"].astype(str).map(_map)

# Age at enrollment  ·  adopt (copy)
h_CLSA["Age At Enrollment (years)"] = raw_CLSA["AGE_NMBR_TRM"]

# Diastolic blood pressure  ·  refine (unit: kPa → mmHg)
h_CLSA["Diastolic Blood Pressure (mmHg)"] = raw_CLSA["BP_DIA_TRM"] * 7.50062 + 0.0

# Diabetes diagnosis  ·  refine (copy)
h_CLSA["Diabetes Mellitus Indicator"] = raw_CLSA["DIA_DX_TRM"]

# Height (cm)  ·  adopt (copy)
h_CLSA["Standing Height (cm)"] = raw_CLSA["HGT_CM_TRM"]
```

## 3. Combine & export

Concatenate the harmonized frames and write the result.

```python
harmonized = pd.concat([h_AoU, h_CLSA], keys=['AoU', 'CLSA'], names=['cohort'])
harmonized.to_csv('harmonized.csv')
harmonized.head()
```
